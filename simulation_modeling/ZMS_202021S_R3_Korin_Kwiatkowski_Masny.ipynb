```python
import numpy as np
import pandas as pd
import scipy as sc
import matplotlib.pyplot as plt
import seaborn as sns
import math
from matplotlib.pyplot import figure
import statistics
plt.style.use('seaborn-whitegrid')
```

##### Przypisanie odpowiednich początkowych wartości parametrom modelu

```python
# global
time_horizon = 90             # w dniach
km_day       = 50             # w kilometrach na dzien
iterations   = 10             # liczba przeprowadzonych iteracji


# own_car
car_cost              = 80000 # koszt zakupu samochodu w zlotowkach
price_drop            = 15    # spadek ceny "po wyjechaniu z salonu"
car_fuel_consumption  = 9     # spalanie wlasnego samochodu na 100 km
fuel_price            = 5.0   # cena paliwa
oc_ac                 = 1000  # koszt ubezpieczenia rocznie
repair_var            = 0.05  # koszt naprawy liczony wyrazony w procencie wartosci samochodu
age_parameter         = 0.001 # spadek wartosci samochodu per dzien 
mu_parameter          = 48    # do rozkladu poissona
fine                  = 5000  # grzywna za parkowanie bez znacznika na parkingu firmy


# carsharing
km_limit_month = 300          # limit km na miesiac w abonamencie
cs_day_cost    = 40           # koszt zakupu abonamentu per dzien
km_cost        = 1            # koszt za km w carsharingu
penalty        = 3            # wspolczynik dodatkowej oplaty za km poza abonamentem (> 1)
crash                 = 5000  # koszt wypadku w carsharingu

# uber
uber_rides = 5                # ile razy zamówiony został uber
uber_init  = 10               # opłata początkowa
uber_km    = 1.8              # koszt km za kilometr (wliczony jest koszt związany z czasem przejazdu)
```

# Model symulacyji <a class="anchor" id="chapter2"></a>

```python
def model(time_horizon, km_day, car_cost, price_drop, car_fuel_consumption, fuel_price, oc_ac, repair_var, age_parameter, mu_parameter, fine, km_limit_month, cs_day_cost, km_cost, penalty, crash, uber_rides, uber_init, uber_km):
    
    own_car = 0
    carsharing = 0
    uber = 0

    # liczba przejechanym km w calym modelowanym okresie dla wlasnego samochodu i carsharingu
    total_km = km_day * time_horizon

    # own_car - wlasny samochod

    # zwiekszenie kosztu o zakup samochodu
    own_car += car_cost
    # zapamietanie poczatkowej wartosci samochodu
    car_value = car_cost
    # spadek wartosci samochodu po wyjechaniu z salonu
    car_value = car_value * ((100 - price_drop) / 100)

    for day in range(1, time_horizon + 1): 
        # zwiekszenie kosztu o koszt zwiazany z liczba przejechanych km dziennie
        own_car += (km_day * car_fuel_consumption / 100) * fuel_price
        # zmniejszenie wartosci samochodu w zwiazku z rosnacym przebiegiem i wiekiem
        car_value -= car_value * age_parameter + km_day * (age_parameter / 10)
        if (day % 365 == 0):
            fuel_price *= 1.015

    # liczba miesiecy w okresie modelowanym
    month_no = math.ceil(time_horizon / 365 * 12) 
    # zmienna odpowiadajaca za prawdopodobienstwo awarii
    a = 1 
    # zmienna zwiekszajaca prawdopodobienstwo kazdej kolejnej awarii
    j = 1 
    # koszt ewentualnych napraw samochodu - rozpatrujemy prawdopodobienstwo awarii raz w kazdym miesiacu
    for months in range(month_no):
        # prawdopodobienstwo awarii dane jest rozkladem poissona
        accident_probability = sc.stats.poisson.cdf(a*j, mu_parameter)
        if accident_probability > np.random.sample():
            # zwiekszenie kosztu, jesli nastapila awaria
            own_car += car_cost * repair_var 
            a = 0
            j += 1
        a += 1

    # liczba szans, w ktorych mogla nastapic grzywna
    fine_chance = fine // 1000
    for k in range(fine_chance):
        # jesli zdarzy sie kontrola to grzywna nastepuje z 10% prawdopodobienstwem
        if np.random.sample() > 0.9:
            # zwiekszenie kosztu samochodu o koszt zwiazany z grzywna
            own_car += fine

    # dodanie kosztu ubezpieczenia w calym modelowanym okresie
    own_car = own_car + (time_horizon // 365 + 1) * oc_ac
    # finalny koszt posiadania wlasnego samochodu pomniejszony o aktualna wartosc samochodu
    own_car = own_car - car_value 


    # carsharing

    # dzien danego miesiaca
    day_no = 0

    for day in range(1, time_horizon + 1):
        
        # obliczenie dnia miesiaca oraz numeru danego miesiaca (zakładamy 30 dniowy miesiac)
        day_no += 1
        single_month_number = day_no / 30
        # wyzerowanie licznika dni po uplynieciu jednego miesiaca
        if single_month_number > 1:
            day_no = 0

        if km_day * day_no <= km_limit_month:
            # zwiekszenie kosztu korzystania z carsharingu od przejechanych km i oplaty abonamentowej
            carsharing += km_day * km_cost + cs_day_cost
        else:
            # zwiekszenie kosztu korzystania z carsharingu od przejechanych km i oplaty abonamentowej z dodatkowym uwzglednieniem kary za przekroczenie limitu dostepnych km
            carsharing += km_day * km_cost * penalty + cs_day_cost
         
    crash_chance = total_km // 1000
    for i in range(crash_chance):
        # wypadek nastepuje z 10% prawdopodobienstwem
        if np.random.sample() > 0.9:
            # zwiekszenie kosztu carsharingu o koszt zwiazany z wypadkiem
            carsharing += crash
            
    # uber

    # koszt korzystania z ubera
    uber = (uber_rides * uber_init + uber_km * km_day) * time_horizon

    # wyniki
    own_car = round(own_car, 2)
    carsharing = round(carsharing, 2)
    uber = round(uber, 2)

    return [own_car, carsharing, uber]
```

```python
def run_model(iterations, time_horizon, km_day, car_cost, price_drop, car_fuel_consumption, fuel_price, oc_ac, repair_var, age_parameter, mu_parameter, fine, km_limit_month, cs_day_cost, km_cost, penalty, crash, uber_rides, uber_init, uber_km):
    m1 = []
    m2 = []
    m3 = []
    for i in range(1, iterations + 1):
        np.random.seed(i)
        tmp = model(time_horizon, km_day, car_cost, price_drop, car_fuel_consumption, fuel_price, oc_ac, repair_var, age_parameter, mu_parameter, fine, km_limit_month, cs_day_cost, km_cost, penalty, crash, uber_rides, uber_init, uber_km)
        m1.append(tmp[0])
        m2.append(tmp[1])
        m3.append(tmp[2])

  # m1 - lista wartosci dla samochodu
  # m2 - lista wartosci dla carsharingu
  # m3 - lista wartosci dla ubera
  # zwracana jest lista, skladajaca sie z 3 list
    return [m1, m2, m3]
```

# Analiza główna <a class="anchor" id="chapter3"></a>

Dla domyślnych wartości, które na początku założył Pan Piotr wyniki prezentują się następująco:

```python
first = run_model(iterations, time_horizon, km_day, car_cost, price_drop, car_fuel_consumption, fuel_price, oc_ac, repair_var, age_parameter, mu_parameter, fine, km_limit_month, cs_day_cost, km_cost, penalty, crash, uber_rides, uber_init, uber_km)
print('Koszt korzystania z własnego samochodu:', np.mean(first[0]), '[odchylenie standardowe:', round(np.std(first[0]),2), ']')
print('Koszt korzystania z carsharingu:', np.mean(first[1]), '[odchylenie standardowe:', np.std(first[1]), ']')
print('Koszt korzystania z ubera:', np.mean(first[2]), '[odchylenie standardowe:', np.std(first[2]), ']')
```

### Jak na łączny koszt eksploatacji własnego samochodu osobowego wpływa parametr początkowej utraty wartości samochodu po wyjechaniu z salonu, współczynnik starzenia się samochodu oraz koszt zakupu samochodu? (analiza 3 parametrów) <a class="anchor" id="section43"></a>

##### Przywrócenie parametrów do wartości początkowych

Rekomendowanym rozwiązaniem jest korzystanie z usług car sharingu dla dystansów poniżej 17 km dziennie, korzystanie z Ubera przy dystansach od 17 do 150 km/d oraz posiadanie własnego samochodu przy pokonywanych dystansach powyżej 150km/d ze względu na niższe koszty.

W przypadku car sharingu ten wzrosrt jest największy, zaś w przypadku własnego samochodu najmniejszy.
